```python
x_values_2014 = eada_2014_df["Total Revenues"]
y_values_2014 = eada_2014_df["Total Expenses"]

slope, intercept, rvalue, pvalue, stderr = linregress(x_values_2014.astype(float), y_values_2014.astype(float))

plt.title('Total Revenues and Total Expenses (2014)')
plt.ylabel('Total Expenses')
plt.xlabel('Total Revenues')
plt.scatter(x_values_2014, y_values_2014)
plt.plot(x_values_2014, intercept + slope * x_values_2014, 'r')
plt.annotate(f'$R^2$: {round(rvalue*rvalue,2)}', (14, 800), c = 'r', fontsize = 16)
plt.show()
```

```python
ttest_2014 = scipy.stats.ttest_ind(eada_2014_df["Total Revenues"],
                                       eada_2014_df["Total Expenses"])

print(ttest_2014)
```
